```python
# connect and close mysql server
def sqlcon(query=0,msg = 'Success'):
    try:
        con = pymysql.connect(host="localhost",user = USERNAME.get(),password = PASSWORD.get(),database = "lib")
        if query == 0:
            return con
        cur = con.cursor()
        cur.execute(query)
        con.commit()
        messagebox.showinfo("Success",msg)   
        con.close()
    except Exception as e:
        messagebox.showinfo("Database Error(sqlcon)",str(e)+"\n"+query)
```

```python
# Add Book Interface

# FRONTEND
def addBook(win):
    
    bodyFrame = Frame(win,bg='white',bd=5,relief='ridge',)
    bodyFrame.place(relx=0,rely=0.15,relwidth=1,relheight=0.7)
    
    
    labeltxt = ('ACC NO. :','BOOK TITLE :','AUTHOR :','PUBLISHER: ','PRICE :') 
    info = (StringVar(),StringVar(),StringVar(),StringVar(),DoubleVar())
    
    heading2 = Label(
        bodyFrame,
        text = 'Enter Book Details...',
        bg = '#ffffff',
        fg = '#2E86C1',
        bd = 3,
        font = ('Courier',18,'bold'),
        relief = 'flat',
    )
    heading2.place(relx = 0,rely = 0,relwidth = 1,relheight = 0.1)
    
    for i in range(5):
        label = Label(
            bodyFrame,
            text = labeltxt[i],
            bg = '#0f0f0f',
            fg = '#ffffff',
            bd = 5,
            font = ('Courier',18,'bold'),
            relief = 'groove',
        )
        label.place(relx = 0.125,rely = 0.15*i+0.125,relwidth = 0.35,relheight = 0.1)
            
    for i in range(5):
        txtBox = Entry(
            bodyFrame,
            bg = '#ffffff',
            fg = '#000000',
            bd = 5,
            font = ('Courier',18,'bold'),
            relief = 'groove',
            justify = 'center',
            textvariable = info[i],
        )
        txtBox.place(relx = 0.525,rely = 0.15*i+0.125,relwidth = 0.35,relheight = 0.1)
    
    addBtn(win,info,bodyFrame)
# BACKEND
   
# submit button to add details into database
def addBtn(win,info,destruct = 1):

    submitbtn = Button(
        footerframe,
        text = 'SUBMIT',
        font = ('Arial',12,'bold'),
        bg = '#16A085',
        fg = '#ffffff',
        activebackground = '#ffffff',
        activeforeground = '#16A085',
    )
    submitbtn.place(relx = 0.444,rely = 0.25,relheight = 0.5, relwidth = 0.1)
    submitbtn.config(command = lambda:[fetchinfo(info)] )
    
    if destruct != 1:
        backBtn(win,destruct)

# fetching information from user and adding into database
def fetchinfo(info):
    try:
        query = "insert into books values"
        query += f"('{info[0].get().upper()}','{info[1].get().upper()}','{info[2].get().upper()}','{info[3].get().upper()}',{info[4].get()})"
        sqlcon(query,'Book Added Successfully')
    except Exception as e:
        messagebox.showinfo('Error in datatype(fetchinfo)',e)

# Back Button
def backBtn(win,frame):
    backbtn = Button(
        footerframe,
        text = 'BACK',
        font = ('Arial',12,'bold'),
        bg = '#16A085',
        fg = '#ffffff',
        activebackground = '#ffffff',
        activeforeground = '#16A085',
        command = lambda:[frame.destroy(),footerframe.destroy(),winfooter(win)]
    )
    backbtn.place(relx = 0.05,rely = 0.25,relheight = 0.5, relwidth = 0.1)
```

```python
# Deletebook

# FrontEnd
def deleteBook(win):
    bodyFrame = Frame(win,bg='white',bd=5,relief='ridge')
    bodyFrame.place(relx=0,rely=0.15,relwidth=1,relheight=0.7)
    
    # What do want to delete heading
    heading2 = Label(
        bodyFrame,
        text = 'What do you want to Delete...',
        bg = '#ffffff',
        fg = '#2E86C1',
        bd = 3,
        font = ('Courier',20,'bold'),
        relief = 'flat',
    )
    heading2.place(relx = 0,rely = 0,relwidth = 1,relheight = 0.1)
    
    #dropdown menu to select colName to Delete From
    col_name = ('ACC_NO','TITLE','AUTHOR','PUBLISHER','PRICE')
    col = StringVar()
    col.set(col_name[0])
    
    dropdown = OptionMenu(bodyFrame,col,*col_name)
    dropdown.place(relx = 0.1,rely = 0.12,relwidth = 0.2,relheight = 0.1)
    
    # enter Book name to delete
    infoBox = Entry(
        bodyFrame,
        bg = '#ffffff',
        fg = '#000000',
        bd = 3,
        font = ('Courier',18,'bold'),
        relief = 'groove',
        justify = 'center',
    )
    infoBox.place(relx = 0.3,rely = 0.12,relwidth = 0.5,relheight = 0.1)
    
    # lists of available books for delete
    goBtn = Button(bodyFrame,text='GO',justify='center',font=('Courier',10))
    goBtn.place(relx = 0.81,rely=0.12,relwidth=0.09,relheight=0.1)
    goBtn.config(
        command = lambda:[
            showList(bodyFrame,0.32,infoBox.get(),col.get()),
            DELBTN(bodyFrame,0.32,infoBox.get(),col.get())
        ]
    )
    
    
    backBtn(win,bodyFrame)

#delete button
def DELBTN(frame,pos,info,col):
    delbtn = Button(
        footerframe,
        text = 'DELETE',
        font = ('Arial',12,'bold'),
        bg = '#16A085',
        fg = '#ffffff',
        activebackground = '#ffffff',
        activeforeground = '#16A085',
    )
    delbtn.place(relx = 0.444,rely = 0.25,relheight = 0.5, relwidth = 0.1)
    delbtn.config(command = lambda:[delinfo(info,col),showList(frame,pos,info,col)] )

    # BACKEND    
# delete button function
def delinfo(info,col):
    try:
        if col == 'Price':
            query = f"delete from books where {col} = {info}"
        elif col == 'AVAILABILITY':
            query = f"delete from books where {col} = {info}"
        else:
            query = f"delete from books where {col} = '{info}'"
        sqlcon(query,f'{info} Deleted Successfully')
    except Exception as e:
        messagebox.showinfo('Error in Database(delinfo)',f'{info} is not available\n'+str(e))
```

```python
# Issuebooks
def issueBook(win):
    bodyFrame = Frame(win,bg='#ffffff',bd=5,relief='ridge')
    bodyFrame.place(relx=0,rely=0.15,relwidth=1,relheight=0.7)
    
    # heading
    heading2 = Label(
        bodyFrame,
        text = 'Student Corner',
        bg = '#ffffff',
        fg = '#2E86C1',
        bd = 3,
        font = ('Courier',20,'bold'),
        relief = 'flat',
    )
    heading2.place(relx = 0,rely = 0,relwidth = 1,relheight = 0.1)
    
    label = Label(bodyFrame,text = 'Student ID :',fg = '#ffffff',bg = '#000000',font=('Courier',13,'bold'),relief='groove')
    label.place(relx = 0.1,rely = 0.11,relwidth = 0.2,relheight=0.09)
    
    studentID = [StringVar(),StringVar(),StringVar()]
    
    studentID1 = Entry(
        bodyFrame,
        bg = '#ffffff',
        fg = '#000000',
        bd = 3,
        font = ('Courier',18,'bold'),
        relief = 'groove',
        justify = 'center',
        textvariable = studentID[0],
    )
    studentID1.place(relx = 0.35,rely = 0.11,relwidth = 0.09,relheight = 0.1)
    studentID1.insert(0,'SSGI')
    label = Label(
        bodyFrame,
        bg = '#ffffff',
        fg = '#000000',
        font = ('Courier',18,'bold'),
        justify = 'center',
        text = '-',
    )
    label.place(relx = 0.45,rely = 0.11,relwidth = 0.05,relheight = 0.09)
    studentID2 = Entry(
        bodyFrame,
        bg = '#ffffff',
        fg = '#000000',
        bd = 3,
        font = ('Courier',18,'bold'),
        relief = 'groove',
        justify = 'center',
        textvariable = studentID[1],
    )
    studentID2.place(relx = 0.50,rely = 0.11,relwidth = 0.1,relheight = 0.09)
    studentID2.insert(0,'2019')
    label = Label(
        bodyFrame,
        bg = '#ffffff',
        fg = '#000000',
        font = ('Courier',18,'bold'),
        justify = 'center',
        text = '-',
    )
    label.place(relx = 0.60,rely = 0.11,relwidth = 0.05,relheight = 0.09)
    studentID3 = Entry(
        bodyFrame,
        bg = '#ffffff',
        fg = '#000000',
        bd = 3,
        font = ('Courier',18,'bold'),
        relief = 'groove',
        justify = 'center',
        textvariable = studentID[2],
    )
    studentID3.place(relx = 0.65,rely = 0.11,relwidth = 0.1,relheight = 0.09)

    stuBtn = Button(bodyFrame,text='GO',justify='center',font=('Courier',15,'bold'),relief='groove')
    stuBtn.place(relx = 0.81,rely=0.11,relwidth=0.09,relheight=0.09)
    stuBtn.config(
        command = lambda:[stuSearch(bodyFrame,0.2,studentID)]
    )
    backBtn(win,bodyFrame)
# frame 2    
def stuSearch(frame,pos,info):
    
    # 1. student information label
    if isinstance(info,str):
        stuId = info
    else:
        stuId = ''
        for i in range(3):
            stuId += info[i].get()
    stuInfoLabel = Label(frame,bg = '#fffeee',fg='#000000',font=('Courier',10),justify = 'center',relief = 'groove')
    stuInfoLabel.place(relx = 0.1,rely = pos+0.02,relwidth = 0.8, relheight = 0.1)
    
    # 2. To show list of books issued frame
    bookInfoFrame = Frame(frame,bg = '#ffffff')
    bookInfoFrame.place(relx=0.1,rely=pos+0.2,relwidth=0.8,relheight=0.78-pos)
    
    # 3. connect to sql to display issued books
    try:
        query = f"select * from student where ID LIKE '{stuId}'"
        con = sqlcon()
        cur = con.cursor()
        cur.execute(query)
        stuInfo = cur.fetchall()
        if len(stuInfo)==0:
            con.close()
            newStu(frame,pos,stuId)
        else:
            stuInfoLabel = Label(frame,bg = '#ffffee',fg='#0000ff',font=('Courier',16,'bold'),justify = 'center',relief = 'groove')
            stuInfoLabel.place(relx = 0.1,rely = pos+0.02,relwidth = 0.8, relheight = 0.1)
            stuInfoLabel.config(text = stuInfo)
        
            query = "SELECT BOOKS.*, BOOKS_ISSUED.DOI "
            query+= "FROM BOOKS,BOOKS_ISSUED "
            query+= "WHERE BOOKS.ACC_NO = BOOKS_ISSUED.ACC_NO "
            query+= f"AND BOOKS_ISSUED.STUDENT_ID IN ('{stuId}')"
            
            cur.execute(query)
            booklists = cur.fetchall()
            con.close()
            if len(booklists)==0:
                booklabel = Label(bookInfoFrame,bg='#ffffff',fg='#000000',font=('courier',20),justify='center')
                booklabel.pack(fill=BOTH,expand=True)
                booklabel.config(text = 'No books Issued')
                
            else:
                i=0
                for book in booklists:
                    booklabel = Label(
                        bookInfoFrame,
                        bg='#fffff0',
                        fg='#000000',
                        font=('courier',13),
                        justify='center',
                        relief='groove',
                        bd=3
                    )
                    booklabel.place(relx=0,rely= i*0.125,relwidth=0.9,relheight=0.125)
                    booklabel.config(text = f"{book[0]}\t{book[1]}\t{book[2]}\t{book[3]}\t{book[4]}\t{book[5]}")
                    
                    # RETURN BUTTON
                    btnlabel = Button(bookInfoFrame,text='RETURN',command=lambda x=book[0]:[retBook(stuId,x,frame,pos)])
                    btnlabel.place(relx=0.9,rely= i*0.125,relwidth=0.1,relheight=0.125)
                    i+=1
            
            # if already 4 books are issued then issue option is not available
            if len(booklists)<4:
                accNo = Entry(frame,bg = '#ffffff',fg='#f00000',font = ('Arial',13,'bold'),justify='center')
                accNo.place(relx = 0.4,rely=0.8,relwidth=0.2,relheight=0.1)
                accNo.insert(0,'Enter Acc No...')
                ISSBTN(frame,pos,stuId,accNo)
            else:
                accNo = Label(frame,bg = '#ffffff',fg='#f00000',font = ('Arial',13,'bold'),justify='center')
                accNo.place(relx = 0.2,rely=0.8,relwidth=0.6,relheight=0.1)
                accNo.config(text='Maximum 4 books can be issued')
            
    except Exception as ex:
        messagebox.showinfo('Software Error(stuSearch)','Error Found In Code\n'+str(ex))

def ISSBTN(frame,pos,ID,bookNo):
    issbtn = Button(
        footerframe,
        text = 'ISSUE',
        font = ('Arial',12,'bold'),
        bg = '#16A085',
        fg = '#ffffff',
        activebackground = '#ffffff',
        activeforeground = '#16A085',
    )
    issbtn.place(relx = 0.444,rely = 0.25,relheight = 0.5, relwidth = 0.1)
    issbtn.config(command = lambda:[ISSBtnFunc(ID,bookNo.get(),frame,pos)] )

def ISSBtnFunc(ID,bookNo,frame,pos):
    try:
        if bookNo.isdigit():
            query = "insert into BOOKS_ISSUED (STUDENT_ID,ACC_NO) "
            query += f"VALUES('{ID}',{bookNo})"
        
            con = sqlcon()
            cur = con.cursor()
            cur.execute(f"select * from books where ACC_NO = {bookNo}")
            book = cur.fetchall()
            con.close()
        
            if len(book)==1:
                sqlcon(query,f'{bookNo} Issued to {ID} Successfully')
                stuSearch(frame,pos,ID)
            else:
                messagebox.showinfo('Check ACC No.',f'Incorrect {bookNo}')
        else:
            messagebox.showinfo('Error',f'Please enter Acc No(DIGITS ONLY)')
    except Exception as e:
        messagebox.showinfo('Error in Database(delinfo)',f'{info} is not available\n'+str(e))

# Book returnBtn function
def retBook(ID,bookNo,frame,pos):
    try:
        query = f"delete from books_issued where acc_no = '{bookNo}'"
        sqlcon(query,f'Book {bookNo} returned Successfully')
        stuSearch(frame,pos,ID)
    except Exception as e:
        messagebox.showinfo('Error in Database(retbook)',f'{bookNo} is not not issued\n'+str(e))
```

```python
# Student Info
def studentInfo(win):
    bodyFrame = Frame(win,bg='white',bd=5,relief='ridge')
    bodyFrame.place(relx=0,rely=0.15,relwidth=1,relheight=0.7)
    
    # dropdown menu
    options = ('ACC NO','STUDENT ID','STUDENT NAME','BRANCH','SEMESTER')
    choice = StringVar()
    choice.set(options[1])
    dropdown = OptionMenu(bodyFrame,choice,*options)
    dropdown.place(relx = 0.1,rely = 0.05,relwidth = 0.15,relheight = 0.1)
    
    inpBox = Entry(
        bodyFrame,
        bg = '#ffffff',
        fg = '#0000ff',
        font = ('Arial',14,'bold'),
        justify = 'center',
    )
    inpBox.place(relx = 0.25,rely = 0.05,relwidth = 0.5,relheight = 0.1)
    
    #Seaarch Button
    searchBtn = Button(
        bodyFrame,
        text = 'SEARCH',
        bg = '#ffffff',
        fg = '#0000ff',
        font = ('Courier',13,'bold'),
    )
    searchBtn.place(relx = 0.75,rely = 0.05,relwidth = 0.15,relheight = 0.1)
    searchBtn.config(command = lambda:[searchBtnFunc(choice.get(),inpBox.get(),bodyFrame,0.15)])
    backBtn(win,bodyFrame)
# button for student information   
def searchBtnFunc(choice,info,frame,pos):
    try:
        infoFrame = Frame(frame)
        infoFrame.place(relx = 0.1,relwidth=0.8,rely=pos+0.02,relheight=0.96-pos)
        if info=='':
            txt = ('S.NO','STUDENT ID','NAME','SEMESTER','BRANCH','PHONE NO.')
            query = "SELECT * from STUDENT"
        
        elif choice=='SEMESTER':
            txt = ('S.NO','STUDENT ID','NAME','SEMESTER','BRANCH','PHONE NO.')
            query = f"SELECT * from STUDENT WHERE SEM = {info}"
        
        elif choice == 'ACC NO':
            txt = ('S.NO','ISSUE DATE','STUDENT ID','NAME','SEMESTER','BRANCH')
            query = "SELECT BOOKS_ISSUED.DOI,STUDENT.ID,STUDENT.NAME,STUDENT.SEM,STUDENT.BRANCH "
            query += "FROM STUDENT,BOOKS_ISSUED "
            query += f"WHERE BOOKS_ISSUED.STUDENT_ID = STUDENT.ID AND BOOKS_ISSUED.ACC_NO LIKE \"%{info}%\""
        
        else:
            Dict = {'STUDENT ID':'ID','STUDENT NAME':'NAME','BRANCH':'BRANCH'}
            txt = ('S.NO','STUDENT ID','NAME','SEMESTER','BRANCH','PHONE NO.')
            query = f"SELECT * FROM STUDENT WHERE {Dict[choice]} LIKE \"%{info}%\""

        con = sqlcon()
        cur = con.cursor()
        cur.execute(query)
        studentFound = cur.fetchall()
        con.close()
        if len(studentFound) == 0:
            messagebox.showinfo('Not Available',f'Book with {choice} {info}\nis Not Issued OR not FOUND')
        else:
            # treeview to show database table            
            table = ttk.Treeview(infoFrame)
            
            table["columns"] = tuple(range(len(txt)))
            table['show']='headings'
            
            s = ttk.Style(infoFrame)
            s.theme_use("clam")
            s.configure(".", font = ('Helvetika',13),rowheight=25) # '.' is for all rows
            s.configure("Treeview.Heading", foreground='#ff0000',font = ('Helvetika',12,'bold'),rowheight=40)
            
            table.column(0,width=5,anchor=CENTER)
            table.heading(0,text=txt[0],anchor=CENTER)
            
            for j in range(1,len(txt)):
                table.column(j,width=20,minwidth=20,anchor=CENTER)
                table.heading(j,text=txt[j],anchor=CENTER)
            
            i=0
            for stu in studentFound:
                table.insert('',i,values=(i+1,stu[0],stu[1],stu[2],stu[3],stu[4]))
                i+=1
            scroll = Scrollbar(infoFrame,orient="vertical",command=table.yview)
            table.config(yscrollcommand=scroll.set)
            scroll.pack(side=RIGHT,fill=Y)
            table.pack(fill=BOTH,expand=True)
    except Exception as e:
        messagebox.showinfo('Error',f'searchBtnFunc has error\n'+str(e))
```

```python
def showList(frame,pos,data,colName,tableName='books'):
    infoFrame = Frame(frame,bg = '#ffffff')
    infoFrame.place(relx=0.1,rely=pos+0.02,relwidth=0.8,relheight=0.96-pos)
    
    try:
        if len(data)==0:
            query = f"select * from {tableName}"
        else:
            query = f"select * from {tableName} where {colName} LIKE \"%{data}%\""
        
        # getting information from database
        con = sqlcon()
        cur = con.cursor()
        cur.execute(query)
        books = cur.fetchall()
        con.close()
        
        if len(books)==0:
            e = Label(infoFrame,bg = '#ffffff',fg='#ff0000',font=('Courier',10),justify = 'center',relief = 'groove')
            e.place(relx = 0,rely = 0,relwidth = 1,relheight = 0.1)
            e.config(text=f"{colName} Not in Database")
        else:
            
            # treeview to show database table
            labeltxt = ('S.NO','ACC NO.','BOOK TITLE','AUTHOR','PUBLISHER','PRICE')
            
            table = ttk.Treeview(infoFrame)
            
            table["columns"] = (0,1,2,3,4,5)
            table['show']='headings'
            
            s = ttk.Style(infoFrame)
            s.theme_use("clam")
            s.configure(".", font = ('Helvetika',13),rowheight=40) # '.' is for all rows
            s.configure("Treeview.Heading", foreground='#ff0000',font = ('Helvetika',14,'bold'),rowheight=40)
            
            table.column(0,width=5,anchor=CENTER)
            table.heading(0,text=labeltxt[0],anchor=CENTER)
            
            for j in range(1,6):
                table.column(j,width=20,minwidth=20,anchor=CENTER)
                table.heading(j,text=labeltxt[j],anchor=CENTER)
            
            i=0
            for book in books:
                table.insert('',i,values=(i+1,book[0],book[1],book[2],book[3],book[4]))
                i+=1
                
            scroll = Scrollbar(infoFrame,orient="vertical",command=table.yview)
            table.config(yscrollcommand=scroll.set)
            scroll.pack(side=RIGHT,fill=Y)
            table.pack(fill=BOTH,expand=True)
        
    except Exception as ex:
        messagebox.showinfo('Software Error(showlist)','Error Found In Code\n'+str(ex))
```
